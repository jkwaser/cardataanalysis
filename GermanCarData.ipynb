```python
# Looks like price is a string with commas and dollar signs.

autos["price"]
```

```python
# Price is currently listed in dollars and has commas. It's also a string. We need to clean it up and convert to an int to be able to get more information out of it.

autos["price"] = autos["price"].str.replace('$','')

autos["price"] = autos["price"].str.replace(',','')


autos["price"] = autos["price"].astype(int)
```

```python
# This looks better.

autos["price"]
```

```python
# That's better. 

autos.describe()
```

```python
# Doing the same for odometer data.

autos["odometer"] = autos["odometer"].str.replace('km','')

autos["odometer"] = autos["odometer"].str.replace(',','')

autos["odometer"] = autos["odometer"].astype(int)
```

```python
autos.describe()
```

```python
# Add some units to the odomoter column name.

autos.rename(columns={'odometer' : 'odometer_km'}, inplace=True)
```

```python
# Let's look at this new price data.

price = autos["price"]
price_uniq = price.unique().shape
print(price_uniq)
price_counts = price.value_counts()
price_counts.sort_index(ascending=True).head(15)
```

```python
price.describe()
```

Price offers a little more granularity. We can see that expensive cars are incredibly rare - with an average price of \\$5900 and 75% of all cars costing under \\$7500.

```python
# Let's look at how common sales are.

date_freqs = autos['date_crawled'].str[:10].value_counts(normalize=True, dropna=False).sort_index()

date_freqs = date_freqs * 100
    
date_freqs.describe()
```

Bear in mind this data comes from a 34-day period, so there are only 34 dates in question. The average day has 2.94 ads crawled, and even the largest day only has 3.86.

```python
date_freqs
```

Our extreme outliers are the last two days of the dataset - again, this seems likely to be a function of the crawler used.

```python
ad_freqs = autos['ad_created'].str[:10].value_counts(normalize=True, dropna=False).sort_index()

ad_freqs = ad_freqs * 100
    
ad_freqs.head(20)
```

```python
ad_freqs.describe()
```

We can see a greater variance here. A lot of this makes a tremendous amount of sense - less cars sold in our given time period will be from much further back. After all, ads posted months prior are much more likely to be have succeeded already.

```python
last_freqs = autos['last_seen'].str[:10].value_counts(normalize=True, dropna=False).sort_index()

last_freqs = last_freqs * 100
    
last_freqs.head(150)
```

This is quite sensible. A large number of ads will have been seen recently. After all, people are looking for cars - very large numbers of ads have been seen in the last few days.

```python
brandlist = autos["brand"].value_counts(normalize=True)
```

```python
brandlist[1:10]

brandy = ["bmw", "opel", "mercedes_benz", "audi", "ford", "renault", "peugeot", "fiat", 'seat']          
```

```python
aggregate_dict = {}

mean_price = {}

for i in brandy:
    selected_rows = autos[autos["brand"] == i]
    selected_row = selected_rows.sort_values("price", ascending = False)
    aggregate_dict[i] = selected_row
    mean_price[i] = selected_row["price"].mean()


        
```

```python
mean_price
```

We see some mostly expected results. Luxury brands like BMW and Mercedes are more expensive, Ford and Renault are less so.

```python
aggregate_dict2 = {}

mean_mileage = {}

for i in brandy:
    selected_rows = autos[autos["brand"] == i]
    selected_row = selected_rows.sort_values("odometer_km", ascending = False)
    aggregate_dict2[i] = selected_row
    mean_mileage[i] = selected_row["odometer_km"].mean()
```

```python
mean_mileage
```

Let's look at the mileage and compare it to the price.

```python
mileage_series = pd.Series(mean_mileage)

price_series = pd.Series(mean_price)
```

```python
mileage_and_price = pd.DataFrame(price_series, columns=['mean_price'])

mileage_and_price["mean_mileage"] = mileage_series

mileage_and_price
```

This is interesting. Mileage is fairly similar among all car brands, but price varies a great deal. This suggests one of the primary determinants of price is branding (or car quality, or some other quality not measured in the dataset numerically) rather than simple mileage
